## Run on images for competition sumission

```python
# When installed using pip
from monk.pytorch_prototype import prototype
```

```python
# Load project in inference mode

gtf = prototype(verbose=0);
gtf.Prototype("Project-Bill", "pytorch-resnext101", eval_infer=True);
```

```python
import os
lst = os.listdir("dataset_bill/test")
```

```python
from tqdm import tqdm
combined = [];
for i in tqdm(range(len(lst))):
    img_name ="dataset_bill/test/" + lst[i];
    predictions = gtf.Infer(img_name=img_name, return_raw=True);
    combined.append([lst[i].split(".")[0], 
                     predictions["raw"][0],
                     predictions["raw"][1],
                     predictions["raw"][2],
                     predictions["raw"][3]])
```

```python
import pandas as pd
df = pd.DataFrame(combined, columns = ['img_id', 'fuel', 'parking', 'restaurant', 'transport'])
```

```python
df.to_csv("submission.csv", index=False)
```

## Data conversion

```python
import pandas as pd
df = pd.read_csv("dataset_bill/train_labels.csv");
```

```python
df.columns
```

```python
df.iloc[0]
```

```python
from tqdm import tqdm
combined = [];
for i in tqdm(range(len(df))):
    img_name = str(df["img_id"][i]) + ".png";
    if(df["fuel"][i]):
        label = "fuel";
    elif(df["parking"][i]):
        label = "parking";
    elif(df["restaurant"][i]):
        label = "restaurant";
    elif(df["transport"][i]):
        label = "transport";
    combined.append([img_name, label]);
```

```python
df = pd.DataFrame(combined, columns = ['Img_Name', 'Label'])
```

```python
df.to_csv("train_labels_modified.csv", index=False)
```

# Training

```python
#Using mxnet-gluon backend 

# When installed using pip
from monk.pytorch_prototype import prototype


# When installed manually (Uncomment the following)
#import os
#import sys
#sys.path.append("monk_v1/");
#sys.path.append("monk_v1/monk/");
#from monk.gluon_prototype import prototype
```

```python
gtf = prototype(verbose=1);
gtf.Prototype("Project-Bill", "sample-experiment-1");
```

```python
gtf.List_Models()
```

```python
gtf.Default(dataset_path="dataset_bill/train",
            path_to_csv="train_labels_modified.csv",
            model_name="resnext50_32x4d", 
            freeze_base_network=False,
            num_epochs=5);
```

```python
gtf.update_batch_size(16);
gtf.update_save_intermediate_models(False);

#Important to reload post updates
gtf.Reload();
```

```python
gtf.Train();
```
